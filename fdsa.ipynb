```python
import yfinance as yf

# Extracting Tesla stock data
tesla_data = yf.download("TSLA", start="2023-01-01", end="2024-01-01")

# Reset index
tesla_data.reset_index(inplace=True)

# Display first five rows
print(tesla_data.head())
```

```python
# Assuming you have already scraped the Tesla revenue data and stored it in the tesla_revenue dataframe

# Display the last five rows
print(tesla_data.tail())
```

```python
import yfinance as yf

# Extracting GameStop stock data
gme_data = yf.download("GME", start="2023-01-01", end="2024-01-01")

# Reset index
gme_data.reset_index(inplace=True)

# Display first five rows
print(gme_data.head())
```

```python
# Assuming you have already scraped the GameStop revenue data and stored it in the gme_revenue dataframe

# Display the last five rows
print(gme_data.tail())
```

```python
import matplotlib.pyplot as plt

# Define a function to make a graph
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Tesla stock graph
make_graph(tesla_data, title='Tesla Stock Price')

```
